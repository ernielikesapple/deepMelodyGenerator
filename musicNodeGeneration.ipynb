generate drop meldy

```
from note_seq.protobuf import music_pb2

drop_melody = music_pb2.NoteSequence()

# Add the notes to the sequence.
drop_melody.notes.add(pitch=60, start_time=0.0, end_time=0.25, velocity=80)
drop_melody.notes.add(pitch=67, start_time=0.25, end_time=1.0, velocity=80)
drop_melody.notes.add(pitch=67, start_time=1.0, end_time=1.25, velocity=80)
drop_melody.notes.add(pitch=64, start_time=1.5, end_time=1.75, velocity=80)
drop_melody.notes.add(pitch=67, start_time=1.75, end_time=2.0, velocity=80)
drop_melody.notes.add(pitch=69, start_time=2.0, end_time=2.5, velocity=80)
drop_melody.notes.add(pitch=67, start_time=2.5, end_time=3.0, velocity=80)
drop_melody.notes.add(pitch=64, start_time=3.0, end_time=3.5, velocity=80)
drop_melody.notes.add(pitch=64, start_time=3.5, end_time=3.75, velocity=80)
drop_melody.notes.add(pitch=62, start_time=3.75, end_time=4.0, velocity=80)
drop_melody.notes.add(pitch=60, start_time=4.0, end_time=4.25, velocity=80)
drop_melody.notes.add(pitch=57, start_time=4.25, end_time=4.58, velocity=80)
drop_melody.notes.add(pitch=57, start_time=4.58, end_time=5.0, velocity=80)
drop_melody.notes.add(pitch=65, start_time=5.0, end_time=6.0, velocity=80)
drop_melody.notes.add(pitch=65, start_time=6.0, end_time=6.25, velocity=80) 
drop_melody.notes.add(pitch=67, start_time=6.25, end_time=7.0, velocity=80)
drop_melody.total_time = 8

drop_melody.tempos.add(qpm=128);

# This is a colab utility method that visualizes a NoteSequence.
note_seq.plot_sequence(drop_melody)

# This is a colab utility method that plays a NoteSequence.
note_seq.play_sequence(drop_melody,synth=note_seq.fluidsynth)

```

convert to midi file

```
# This creates a file called `drop_sample_output.mid`, containing the drop solo we've been using.
note_seq.sequence_proto_to_midi_file(drop_melody, 'drop_sample_output.mid')

# This is a colab utility method to download that file. In your Python script, you 
# would just write it to disk.
files.download('drop_sample_output.mid')
```
